```python
import numpy as np
import pandas as pd
```

```python
# random seed
np.random.seed(42)
set1no = 900
set2no = 100
```

```python
df1 = pd.DataFrame( {
    "feature_1": np.random.normal(loc=0 , scale =1 , size = set1no),
    "feature_2": np.random.normal(loc=0 , scale =1 , size = set1no),
    "target": [0]*set1no
})

df2 = pd.DataFrame( {
    "feature_1": np.random.normal(loc=0 , scale =1 , size = set2no),
    "feature_2": np.random.normal(loc=0 , scale =1 , size = set2no),
    "target": [1]*set2no
})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
```

```python
df
```

```python
df["target"].unique()
```

```python
df["target"].value_counts()
```

```python
# upsampling ==> upsample minority
# downsampling ==> downsample majority
```

```python
df_minority = df[df["target"]==1]
```

```python
df_minority
```

```python
df_majority = df[df["target"]==0]
```

```python
df_majority
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsample=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
```

```python
df_minority_upsample
```

```python
df_minority_upsample.shape
```

```python
df_minority_upsample.head()
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsample])
```

```python
df_upsampled
```

```python
df_upsampled["target"].value_counts()
```

```python
resample(df_majority,replace=True,n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsample = resample(df_majority,replace=True,n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsample
```

```python
df_majority_downsample.shape
```

```python
df_majority_downsample.head()
```

```python
df_majority_downsample["target"].value_counts()
```

```python
df_downsample = pd.concat([df_minority,df_majority_downsample])
```

```python
import imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
```

```python
(X,y)=oversample.fit_resample(df[["feature_1","feature_2"]],df["target"])
```

```python
X
```
