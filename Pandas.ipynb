```python
df = pd.read_csv('8-apply_function_data.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.columns
```

```python
def salary_category(salary):
  if salary < 50000:
    return "Low"
  elif 50000 <= salary < 80000 :
    return "Medium"
  else :
    return "High"
```

```python
df['Salary_Category'] = df['Salary'].apply(salary_category)
```

```python
df
```

```python
def erformance_score(score):
  if score < 2:
    return "Low"
  elif 2 <= score < 5 :
    return "Medium"
  else :
    return "High"
```

```python
df['Performance_Score_Category'] = df['Performance_Score'].apply(erformance_score)
```

```python
df
```

```python
df['Formatted_Name'] = df['Name'].apply(lambda x : x.replace("_"," "))
```
