```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('6-bank_customers.csv')
```

```python
df.columns
```

```python
# Scenario: Predicting whether a customer will subscribe to a term deposit (bank marketing use case) == Bir müşterinin vadeli mevduat hesabı açıp açmayacağını tahmin etmek (banka pazarlama kullanım örneği)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop('subscribed',axis=1)
y = df['subscribed']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size=0.25, random_state=(42))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
y_pred = log_model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.metrics import confusion_matrix
score = accuracy_score(y_pred ,y_test)
print("score : ",score)
print(classification_report(y_pred,y_test))
print("confusion matrix \n",confusion_matrix(y_pred,y_test))
```

```python
#hyperparameter tuning
```

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("score: " , score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n" , confusion_matrix(y_pred,y_test))
```
