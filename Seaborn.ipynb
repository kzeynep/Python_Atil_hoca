```python
import matplotlib.pyplot as plt
```

```python
x = [ 1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x,y,color="blue",marker ="o",linestyle ="--")
plt.title ("Basit çizgi grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(x,y,color ="red",linewidth=2,marker ="s",label="Veri-1")
plt.legend()
plt.title("Grafik Başlığı",fontsize = 14,fontweight="bold")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.grid(alpha=0.4)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
# dağılım grafiği
sns.scatterplot(data=df,x="sepal_length",y="petal_length",hue="species",style="species")
plt.title("İris Çiçeği Özellik Dağılımı")
plt.show()
```

```python
# histogram ve yoğunluk grafiği
sns.histplot(data=df,x="sepal_length",kde =True,bins =20,color="skyblue")
plt.title("Sepal Length Histogramı ")
plt.show()
```

```python
#boxplot (kutu grafiği)
sns.boxplot(data = df,x="species",y="petal_length",palette="Set2")
plt.title("Petal Lenght'in Türlere göre dağılımı")
plt.show()
```

```python
#pairplot (tüm değişkenler arası ilişkiler)
sns.pairplot(data=df,hue="species",diag_kind="kde")
plt.show()
```

```python
#korelasyon ısı haritası
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",center=0)
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
sns.set_style("white")
sns.set_style("darkgrid")
sns.set_style("whitegrid")
sns.set_style("ticks")
```

```python
fig,axes =plt.subplots(1,2,figsize=(12,5))
sns.histplot(df["sepal_width"],kde=True,ax=axes[0])
sns.boxplot(x=df["species"],y=df["petal_width"],ax=axes[1])
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

y_true = [0, 1, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```
