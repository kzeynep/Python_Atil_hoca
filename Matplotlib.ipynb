```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Test Grafiği")
plt.show()
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
weight_list = [20,60,80,85,86,87,70,90,95,90]
plt.plot(age_list , weight_list , "r")
plt.title("Age-Weight Graph")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show() #grafiği çizdirmek istediğimiz zaman kullandığımız kod
```

```python
np_age_list = np.array(age_list)
```

```python
np_age_list
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2,"g*-")
plt.show()
```

```python
plt.plot(numpy_arr2, numpy_arr1,"b*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1,"g--")
plt.show()
#
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3]) # grafiği ölçeklendirmek gibi
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Label")
figure_axes.set_ylabel("Y Label")
figure_axes.set_title("Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1 , numpy_arr2 **2, label = "numpy array ** 2")
new_axes.plot(numpy_arr1 , numpy_arr2 **3, label = "numpy array ** 3")
new_axes.legend(loc=6)
plt.show()
```
